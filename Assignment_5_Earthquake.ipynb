## EARTHQUAKE DATA ASSIGNMENT

### INTRODUCTION

- Earthquakes happen when the earth’s tectonic plates (pieces of the earth’s crust & uppermost mantle) shift together, causing friction & splitting the earth’s surface.
- An earthquake’s power (magnitude) can change from those with such low magnitude that they cannot be felt, to those with such high magnitudes that can destroy entire cities/regions.
- Earthquakes are measured by magnitude. This is a quantitative, mathematical calculation to measure the energy released at the source of an earthquake. 
- On the other hand, earthquake intensity is the qualitative measurement of the strength of shaking produced by an earthquake.


### DEFINITIONS

There are various scales for measuring the magnitude / power of an earthquake. For example:
1. Surface wave magnitude scale (MS) is based on the maximum amplitude of the surface
2. Body wave magnitude scale (MB) is calculated from the body waves
3. Moment magnitude scale (MW) is measured over the broad range of frequencies present in the earthquake wave spectrum rather than the single frequency sample that the other magnitude scales use. This is the most widely used scale.
4. Richter magnitude scale (ML) is determined from the logarithm of the amplitude of waves recorded by seismographs.

N.B. It is important to note that each of these scales takes different variables/factors & uses different mathematical formulas to calculate & measure magnitude, so for example, to convert ML into MW or MB is not as straightforward as converting, say, inches into cenimeters.
